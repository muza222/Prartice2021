```python
classes = {'тояни' : 0, 'месте' : 0, 'stoyat': 0,
           'одьб' : 1, 
           'бег' : 2, 
           'елоси' : 3, 
           'естн' : 4, 'одъ' : 4, 'stairs' : 4,
           'втомо' : 5, 'ашин' : 5, 
           'метро' : 6, 'лектро' : 7, 
           'амок' : 8}

answers = {'стояние' : 0, 'ходьба' : 1, 'бег' : 2, 'велосипед' : 3, 'лестница' : 4, 
           'автомобиль' : 5, 'метро' : 6, 'электросамокат' : 7, 'самокат' : 8}

# стояние, ходьба, велосипед, лестница, бег, подъем, подъём, автомобиль, электросамокат, машина, метро
```

```python
def get_action(name):
    for act in sorted(classes, key=lambda x : len(x), reverse=True): 
        if name.lower().find(act) != -1:
            return classes[act]
    return -1
```

```python
def preprocess_data(df, comma=True, a=-1, b=-1):
    
# обрезаем начало и конец трека
    if a == -1:
        st = 1250
    else:
        st = a
            
    if b == -1:
        end = -1250
    else:
        end = b

# приводим данные с Iphone к нужному формату
    if comma is True:
        df = df.applymap(lambda x: str(x).replace(',','.'))
        
        try:
            df['gFx'] = df['gFx'].astype(float)
            df['gFy'] = df['gFy'].astype(float)
            df['gFz'] = df['gFz'].astype(float)
        except KeyError:
            return -1
    return df[st:end]
```

```python
data = []
for df in tq.tqdm(Data):
    data.append(preprocess_data(df))
```

```python
from scipy.stats import iqr
```

```python
def process_data(df):
    ans = []
    try:
        Acc = (df.gFx ** 2 + df.gFy ** 2 + df.gFz ** 2) ** 0.5
        amplitude = Acc.max() - Acc.min()
        ans.append(amplitude)

        ans.append(np.std(Acc))

        ans.append(np.sum(df.gFx))
        ans.append(np.sum(df.gFy))
        ans.append(np.sum(df.gFz))

        ans.append(np.sum(Acc))

        ans.append(np.median(Acc))
        ans.append(np.mean(Acc))


        ans.append(np.percentile(Acc, 75))
        ans.append(np.percentile(Acc, 90))

        ans.append(iqr(df.gFx)) # разность 75 и 25 перцентилей
        ans.append(iqr(df.gFy))
        ans.append(iqr(df.gFz))

        ans.append(np.percentile(df.gFx, 25))
        ans.append(np.percentile(df.gFx, 50))
        ans.append(np.percentile(df.gFx, 75))

        ans.append(np.percentile(df.gFy, 25))
        ans.append(np.percentile(df.gFy, 50))
        ans.append(np.percentile(df.gFy, 75))

        ans.append(np.percentile(df.gFz, 25))
        ans.append(np.percentile(df.gFz, 50))
        ans.append(np.percentile(df.gFz, 75))
    finally:
        return ans
```

```python
features = []
for df, target in tq.tqdm(zip(data,Target)):
    tmp = process_data(df)
    tmp.append(get_action(target))
    features.append(tmp)
```

```python
train = pd.DataFrame(features).rename(columns={0:'ampl', 
                                            1:'std', 
                                            2:'sum_x',
                                            3:'sum_y',
                                            4:'sum_z',
                                            5:'sum',
                                            6:'median',
                                            7:'mean',
                                            8:'perc75',
                                            9:'perc90',
                                            10:'iqr_x',
                                            11:'iqr_y',
                                            12:'iqr_z',
                                            13:'perc25_x',
                                            14:'perc50_x',
                                            15:'perc75_x',
                                            16:'perc25_y',
                                            17:'perc50_y',
                                            18:'perc75_y',
                                            19:'perc25_z',
                                            20:'perc50_z',
                                            21:'perc75_z',
                                            22:'target'})
```

```python
from sklearn.model_selection import train_test_split
X_train_cat, X_validation_cat, y_train_cat, y_validation_cat = train_test_split(train_x, train_y,
                                                                                train_size=0.70, random_state=42,
                                                                                stratify=train_y)
```

```python
model = CatBoostClassifier(iterations=10000,
                         learning_rate=0.02,
                         eval_metric='Accuracy',
                         loss_function="MultiClass",
                         use_best_model=True,
                         early_stopping_rounds=1000,
                         random_strength=1,
                         #task_type='GPU'
                         )
```

```python
model.fit(
    X_train_cat, y_train_cat,
    eval_set=(X_validation_cat, y_validation_cat),
    verbose=500
)
```
